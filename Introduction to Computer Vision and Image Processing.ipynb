# Introduction to Computer Vision and Image Processing

<b>Image Classification</b> : It is process of taking image as input and outputting label of image

<b>Uses of Image Classification</b>:
- Organize photo albums on smart devices 
- Augment Medical professional like helping them identifying anamolies
- Identify images around self-driving cars

<b>Challenges of Image Classification tasks </b>:
- change in viewpoint
- change of illumuniation
- Deformation(Different poses)
- Occlusion(Obstacles on object)
- Background clutter(Clutter in an image is defined as the amount of time-independent back- ground texture which is similar to the target in size, shape, and orientation)

Linear classifier tries to find decision boundary(wx+b =z)

Logistic Regression : It uses sigmoid function (1/1+e^-(z))

Epoch is training/iterating over entire dataset. Iteration = train size/batch size, so iteration is on sample of data. for every iteration we take a sample data and find loss.

Softmax is multi class classification with multiple decision boundaries.

Stochastic Gradient Descent which typically means that the batch size is set to 1, we use Mini-Batch Gradient Descent for more than batch size>1

Classification of image depends on relationship between pixels, so traditional classifiers dosen't work.

We use feature of gradients for getting image features instead of colour etc.

<b>Histogram of Oriented Gradients(HOG)</b>: 
It uses gradient orientations in localized regions of image. It generates historgram for each localized regions
